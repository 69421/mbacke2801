### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
## Write your code here

# Nettoyer : enlever les valeurs manquantes pour la compensation et l'âge
df_clean = df.dropna(subset=['ConvertedCompYearly', 'Age'])

# Calculer la médiane de la compensation par groupe d'âge
median_comp_by_age = df_clean.groupby('Age')['ConvertedCompYearly'].median().reset_index()

# Trier les groupes d'âge pour l'affichage
median_comp_by_age = median_comp_by_age.sort_values('Age')

# Tracer le graphique en ligne
plt.figure(figsize=(10,6))
plt.plot(median_comp_by_age['Age'], median_comp_by_age['ConvertedCompYearly'], marker='o')
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (USD)')
plt.xticks(rotation=30, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
## Write your code here

# Nettoyer : enlever les valeurs manquantes pour la compensation et l'âge
df_clean = df.dropna(subset=['ConvertedCompYearly', 'Age'])

# Définir les groupes d'âge à inclure (ceux contenant '25', '30', '35', '40', '45')
age_range = [age for age in df_clean['Age'].unique() if any(x in str(age) for x in ['25', '30', '35', '40', '45'])]
df_range = df_clean[df_clean['Age'].isin(age_range)]

# Calculer la médiane de la compensation par groupe d'âge sélectionné
median_comp_by_age = df_range.groupby('Age')['ConvertedCompYearly'].median().reset_index()

# Trier les groupes d'âge pour l'affichage
median_comp_by_age = median_comp_by_age.sort_values('Age')

# Tracer le graphique en ligne
plt.figure(figsize=(8,5))
plt.plot(median_comp_by_age['Age'], median_comp_by_age['ConvertedCompYearly'], marker='o', color='green')
plt.title('Median Yearly Compensation (Ages 25 to 45)')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (USD)')
plt.xticks(rotation=30, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
## Write your code here

# On suppose que la colonne d'expérience s'appelle 'WorkExp' (ou similaire)
# Vérifie le nom exact si besoin : print(df.columns)

# Nettoyer : enlever les valeurs manquantes pour l'expérience et la satisfaction
df_clean = df.dropna(subset=['WorkExp', 'JobSatPoints_6'])

# Convertir l'expérience en numérique si besoin
df_clean.loc[:, 'WorkExp'] = pd.to_numeric(df_clean['WorkExp'], errors='coerce')
df_clean = df_clean.dropna(subset=['WorkExp'])

# Calculer la moyenne ou la médiane de la satisfaction par niveau d'expérience
job_sat_by_exp = df_clean.groupby('WorkExp')['JobSatPoints_6'].mean().reset_index()

# Trier par expérience croissante
job_sat_by_exp = job_sat_by_exp.sort_values('WorkExp')

# Tracer le graphique en ligne
plt.figure(figsize=(10,6))
plt.plot(job_sat_by_exp['WorkExp'], job_sat_by_exp['JobSatPoints_6'], marker='o', color='purple')
plt.title('Job Satisfaction (JobSatPoints_6) by Experience Level')
plt.xlabel('Years of Experience')
plt.ylabel('Average Job Satisfaction')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here

# Nettoyer : enlever les valeurs manquantes pour l'expérience et la compensation
df_clean = df.dropna(subset=['WorkExp', 'ConvertedCompYearly'])

# Convertir l'expérience en numérique si besoin
df_clean.loc[:, 'WorkExp'] = pd.to_numeric(df_clean['WorkExp'], errors='coerce')
df_clean = df_clean.dropna(subset=['WorkExp'])

# Calculer la médiane de la compensation par niveau d'expérience
comp_by_exp = df_clean.groupby('WorkExp')['ConvertedCompYearly'].median().reset_index()

# Trier par expérience croissante
comp_by_exp = comp_by_exp.sort_values('WorkExp')

# Tracer le graphique en ligne
plt.figure(figsize=(10,6))
plt.plot(comp_by_exp['WorkExp'], comp_by_exp['ConvertedCompYearly'], marker='o', color='teal')
plt.title('Median Yearly Compensation by Experience Level')
plt.xlabel('Years of Experience')
plt.ylabel('Median Compensation (USD)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here

# Nettoyer : enlever les valeurs manquantes pour l'expérience et la satisfaction
df_clean = df.dropna(subset=['WorkExp', 'JobSatPoints_6'])

# Convertir l'expérience en numérique si besoin
df_clean.loc[:, 'WorkExp'] = pd.to_numeric(df_clean['WorkExp'], errors='coerce')
df_clean = df_clean.dropna(subset=['WorkExp'])

# Calculer la moyenne de la satisfaction par niveau d'expérience
job_sat_by_exp = df_clean.groupby('WorkExp')['JobSatPoints_6'].mean().reset_index()

# Trier par expérience croissante
job_sat_by_exp = job_sat_by_exp.sort_values('WorkExp')

# Tracer le graphique en ligne
plt.figure(figsize=(10,6))
plt.plot(job_sat_by_exp['WorkExp'], job_sat_by_exp['JobSatPoints_6'], marker='o', color='orange')
plt.title('Job Satisfaction (JobSatPoints_6) Across Experience Levels')
plt.xlabel('Years of Experience')
plt.ylabel('Average Job Satisfaction')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
